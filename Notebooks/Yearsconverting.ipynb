```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tmdb_5000_movies.csv")
```

```python
df.head()
```

```python
df["release_date"][0][0:4]
```

```python
df.tail(2)
```

```python
df['release_date'][0:2]
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['year'] = df['release_date'].dt.year
```

```python
df['year']=df['year'].fillna(0)
df['year']=df['year'].astype(int)
```

```python
year_id=df[["title","id","vote_average",'year']]
```

```python
year_id
```

```python
sorted(year_id['year'].unique())
```

```python
import pickle
pickle.dump(year_id.to_dict(),open('yearslist.pkl','wb'))
```
